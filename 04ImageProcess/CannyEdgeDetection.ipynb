```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

There are some tips about Canny Edge Detection
1. Use 5x5 Gaussian filter to remove the Gaussian noise in the image.
1. Calculate the horizontal and vertical gradient ,
based on which to calculate the grad and direction
$$G = \sqrt{G_x^2 + G_y^2}$$
$$\theta = tan^{-1}(\frac{G_y}{G_x})$$
1. Suppose point A is on the edge, B and C which are connected to A in
2 sides are normal to the edge.
That means the line BC is vertical to the edge. So A is possible to
be on the edge only when grad of B and grad of C are all smaller than
grad of A. That means grad A is local maximum. Otherwise, A is definitely not
on the edge
1. make 2 threshold maxVal and minVal. When Grad is greater than maxVal, the
point is definitely on the edge, and not when smaller than minVal. When Grad of
a point is between maxVal and minVal, at the mean time
connects to a point whose Grad is greater than maxVal, it will be
seen as on the edge.

```python
img = cv.imread('../img/football.jpg', 0)
# second and third arguments are minVal and maxVal.
edges = cv.Canny(img, 100, 200)
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Edge'), plt.xticks([]), plt.yticks([])

plt.show()


```
